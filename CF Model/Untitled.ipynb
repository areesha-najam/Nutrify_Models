```python
# Fill in random values from 1 to 10 for each row with a step of 5
idx = []
for i in range(0, len(df1), 5):
    idx.append(i)
    row_values = np.random.randint(1, 11, size=len(df1.columns)-1)  # Generate random values
    df1.iloc[i, 1:] = row_values  # Assign the random values to the current row with a step of 5


# Copy the selected rows to the following 4 rows
for i in idx:
    for j in range(i+1, i+5):
        df1.iloc[j] = df1.iloc[i]
        
df1
```
