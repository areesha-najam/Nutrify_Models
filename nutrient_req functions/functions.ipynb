```python
weight_kg = int(input("Enter your weight in KG: \n"))
height_cm = int(input("Enter your height in cm: \n"))
age = int(input("Enter your age in years: \n"))
isMale = str(input("Are you male? (y/n)"))
activity = input("How active are you? 1 for sedentary, 2 for lightly active, 3 for moderately active, 4 for very active, and 5 for extremely active: ")
goal_weight = float(input("Enter your goal weight in pounds: "))
calories_consumed_per_day = float(input("Enter the number of calories consumed per day: "))

weight_pounds = weight_kg * 2.205
height_inches = height_cm / 2.54

if isMale == "y":
    isMale = True
elif isMale == "n":
    isMale = False
else:
    print("Error")
    quit()

# Bmr: Basal Metabolic Rate: number of calories you burn as your
#  body performs basic (basal) life-sustaining function
#https://www.garnethealth.org/news/basal-metabolic-rate-calculator#:~:text=Your%20basal%20metabolism%20rate%20is,4.330%20x%20age%20in%20years)
def calcBmrHB(): #checked function 
    if isMale:
        bmr =  88.362 + (13.397 * weight_kg) + (4.799 * height_cm) - (5.677 * age)

    else:
        bmr = 447.593 + (9.247 * weight_kg) + (3.098 * height_cm) - (4.330 * age)        
    bmr = round(bmr)
    return(bmr) 

# bmi: body mass index 
def BMI(): #DONE CHECKING FUNCTION 
    bmi = weight_pounds / pow(height_inches,2)*703
    return round(bmi,1)

# # Obesity class 
bmi = BMI()
print(bmi,"bmi")
def obesity_class(bmi):
    if bmi<16:
        obesity= 'Severely Underweight'
    elif bmi<18.5:
        obesity=  'Underweight'
    elif bmi<25.1:
        obesity=  'Normal Weight'
    elif bmi<30.1:
        obesity=  'Overweight'
    elif bmi<35.1:
        obesity=  "Class I Obesity"
    elif bmi<40:
        obesity=  "Class II Obesity"
    elif bmi<50:
        obesity=  "Class III Obesity"
    elif bmi<60:
        obesity=  'Class IV Obesity'
    else :
        obesity=  'Class V Obesity'
    return obesity

#standardized version: to lose 1 pound, one needs to cut down 3500 calories in total
CALORIES_PER_POUND = 3500


harrisBMR = calcBmrHB()
print(harrisBMR,"BMR")
def TDEE(harrisBMR):
    if activity == str(1):
        total = harrisBMR * 1.2
    elif activity == str(2):
        total = harrisBMR * 1.375
    elif activity == str(3):
        total = harrisBMR * 1.55
    elif activity == str(4):
        total = harrisBMR * 1.725
    elif activity == str(5):
        total = harrisBMR * 1.9
    return total

daily_calorie_needs = TDEE(harrisBMR)
print(daily_calorie_needs,"goal calorie per day")
# Calculations
def day_to_reach_goal_weight():
    print(weight_pounds,"your weight in pounds")
    weight_to_lose = goal_weight - weight_pounds
    print(weight_to_lose,"weight to lose (in pounds)")
    calories_needed_to_lose_weight = weight_to_lose * CALORIES_PER_POUND
    print(calories_needed_to_lose_weight,"calories_needed_to_lose_weight")
    calorie_deficit_per_day = daily_calorie_needs - calories_consumed_per_day
    
    print(calorie_deficit_per_day,"calorie_deficit_per_day")
    days_to_reach_goal_weight = calories_needed_to_lose_weight / calorie_deficit_per_day
    return calorie_deficit_per_day,days_to_reach_goal_weight
# Output
output = day_to_reach_goal_weight()
print(f"It will take " + str(round(output[1],2)) + " days to reach your goal weight if you eat " + str(round(output[0])) + " less calories than your current consumption")

```

```python
def carbs_intake(gender, age):
    ans = 0
    a =  df.loc[df['Category'] == gender]
    for index, row in a.iterrows():
        if int(age) > int(row['Start_Age']) and int(age) <= int(row['End_Age']):
            ans = row['Carbohydrates(g)']
    return ans
print("Carbs (g) required:", carbs_intake("Women", 30), "(g)")
```

```python
def protein_intake(gender, age):
    ans = 0
    a =  df.loc[df['Category'] == gender]
    for index, row in a.iterrows():
        if int(age) > int(row['Start_Age']) and int(age) <= int(row['End_Age']):
            ans = row['Protein (g )']
    return ans
print("Protein (g) required:", protein_intake("Women", 30), "(g)")
```
